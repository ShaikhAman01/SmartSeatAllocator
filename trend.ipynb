```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load preprocessed data
data = pd.read_csv('../data/numeric_data.csv')
```

```python
# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# Function to save visualizations in the specified folder
def save_plot(fig, filename):
    fig.savefig(f"../visualizations/{filename}", bbox_inches="tight")
```

```python
# 1. Trend Analysis: Opening and Closing Ranks Over the Years
def plot_trend_analysis(data):
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=data, x="Year", y="Closing_Rank", hue="Gender", marker="o", ci=None)
    plt.title("Trend Analysis of Closing Ranks Over the Years by Gender")
    plt.xlabel("Year")
    plt.ylabel("Closing Rank")
    save_plot(plt, "trend_analysis.png")
    plt.show()
```

```python
# 2. Institute and Program Analysis
def plot_institute_program_analysis(data):
    plt.figure(figsize=(14, 8))
    sns.barplot(data=data, x="Institute", y="Closing_Rank", hue="Academic_Program_Name", ci=None)
    plt.title("Institute and Program Analysis: Closing Ranks")
    plt.xlabel("Institute")
    plt.ylabel("Closing Rank")
    plt.xticks(rotation=45, ha="right")
    save_plot(plt, "institute_program_analysis.png")
    plt.show()
```

```python
# 3. Category and Quota Analysis
def plot_category_quota_analysis(data):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data, x="Seat", y="Closing_Rank", hue="Disability")
    plt.title("Category and Quota Analysis by Disability Status")
    plt.xlabel("Seat Type")
    plt.ylabel("Closing Rank")
    save_plot(plt, "category_quota_analysis.png")
    plt.show()
```

```python
# Run all plots
if __name__ == "__main__":
    plot_trend_analysis(data)
    plot_institute_program_analysis(data)
    plot_category_quota_analysis(data)
```
