```python
import pandas as pd
```

```python
data = pd.read_csv( '../data/numeric_data.csv')
data
```

```python
data['Opening_Rank'] = data['Opening_Rank'].replace(0, pd.NA)

result = data.groupby(['Institute', 'Year']).agg(
    min_opening_rank=('Opening_Rank', 'min'),
    max_closing_rank=('Closing_Rank', 'max')
).reset_index()
result
result.to_csv('../data/Institute_yearly_data.csv', index=False)
```

```python
# data['Opening_Rank'] = data['Opening_Rank'].replace(0, pd.NA)

result2 = data.groupby(['Gender', 'Year']).agg(
    min_opening_rank=('Opening_Rank', 'min'),
    max_closing_rank=('Closing_Rank', 'max')
).reset_index()
result2.to_csv('../data/Gender_yearly_data.csv', index=False)
result2
```
